```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,int(1e5))
y = 10-x + np.random.normal(scale=1e-3,size=len(x))
y += -np.exp(-(x-5)**2/0.001**2)
y += 300*(x-2)*np.exp(-(x-2)**2/0.001**2)
np.savetxt("example2.dat", np.array([x,y]).T)
```
